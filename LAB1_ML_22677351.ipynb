### Bài 2: Gradient Descent

2.1 Giảm độ dốc thủ công cho một hàm đơn giản 

- Triển khai giảm độ dốc để giảm thiểu cho hàm $ f(x)=x^2 $ theo cách thủ công. 
- Yêu cầu: 
1. Khởi tạo x=10, tốc độ học α=0,1 
2. Thực hiện 10 lần lặp lại của phương pháp giảm dần độ dốc theo cách thủ công 
(cập nhật x bằng $ x=x−\alpha * \delta f(x). $ ) 
3. Vẽ đồ thị các giá trị của x và f(x) qua các lần lặp lại. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def grad (x):
    return 2*x  # f'_x
def f_x (x):
    return x**2  # f_x

def my_GD (alpha, x0):
    x = [x0]
    for i in range (10):
        x_new = x[-1] - alpha * grad (x[-1])
        # điều kiện hội tụ
        if abs (grad(x_new)) < 1e-3:
            break
        x.append (x_new)

    return (x,i+1)

(x1,i1) = my_GD (.1, 10) # cho alpha = 0.1, x = 10
print('Solution x1 = %f, f_x = %f, obtained after %d iterations'%(x1[-1], f_x(x1[-1]), i1))

def display(x_vals):  # Duyệt qua danh sách các giá trị x
    j = []
    for k in x_vals:
        y = 2 * k * k + k  # Tính giá trị y dựa trên công thức
        j.append(y)
    return j


j = display(x1)

# Vẽ đồ thị
plt.plot(x1, j)
plt.xlabel('Giá trị x')
plt.ylabel('Giá trị của y')
plt.show()
```

```python
# Chạy Gradient Descent
alpha = 0.1
x0 = 10

# Visualize the Gradient Descent
# Lấy giá trị của x qua các lần lặp
x_vals, _ = my_GD(alpha, x0)
f_vals = [f_x(x) for x in x_vals]

# Đồ thị giá trị của x qua các lần lặp
plt.subplot(1, 2, 1)
plt.plot(range(len(x_vals)), x_vals, marker='o', label='x values')
plt.title('X values over iterations')
plt.xlabel('Iterations')
plt.ylabel('x')
plt.grid(True)
plt.legend()

# Đồ thị giá trị của f(x) qua các lần lặp
plt.subplot(1, 2, 2)
plt.plot(range(len(f_vals)), f_vals, marker='o', label='f(x) values', color='orange')
plt.title('f(x) values over iterations')
plt.xlabel('Iterations')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

### 2.2: Hồi quy tuyến tính với phương pháp giảm dần độ dốc 
- Vẽ đường thẳng y=mx+c bằng phương pháp giảm dần độ dốc. 
- Yêu cầu: 
1. Tạo dữ liệu tổng hợp. 
2. Khởi tạo m và c ngẫu nhiên. 
3. Tính toán độ dốc và cập nhật m và c theo từng lần lặp lại. 
4. Vẽ các điểm dữ liệu và đường hồi quy.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Tạo dữ liệu tổng hợp
np.random.seed(42)
X = np.linspace(0, 10, 100)  # 100 điểm dữ liệu từ 0 đến 10
y = 3 * X + 7 + np.random.randn(100)  # y = 3x + 7 + nhiễu

# 2. Khởi tạo m và c ngẫu nhiên
m = np.random.rand()  # Hệ số góc ban đầu
c = np.random.rand()  # Hệ số chặn ban đầu

# Các siêu tham số
learning_rate = 0.01
epochs = 1000  # Số lần lặp

# Hàm tính mất mát
def compute_loss(X, y, m, c):
    y_pred = m * X + c
    return np.mean((y - y_pred) ** 2)  # MSE

# 3. Thuật toán giảm dần độ dốc
loss_history = []
for epoch in range(epochs):
    # Dự đoán giá trị y
    y_pred = m * X + c
    
    # Tính gradient
    dm = -2 * np.sum((y - y_pred) * X) / len(X)  # Gradient của m
    dc = -2 * np.sum(y - y_pred) / len(X)        # Gradient của c
    
    # Cập nhật m và c
    m -= learning_rate * dm
    c -= learning_rate * dc
    
    # Tính và lưu giá trị mất mát
    loss = compute_loss(X, y, m, c)
    loss_history.append(loss)
    
    # In mất mát mỗi 100 epochs
    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")

# 4. Vẽ các điểm dữ liệu và đường hồi quy
plt.figure(figsize=(10, 6))

# Vẽ các điểm dữ liệu
plt.scatter(X, y, color='blue', label='Data Points')

# Vẽ đường hồi quy
y_final = m * X + c
plt.plot(X, y_final, color='red', label=f'Regression Line (y={m:.2f}x + {c:.2f})')

# Trang trí đồ thị
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression using Gradient Descent')
plt.legend()
plt.grid()
plt.show()

# 5. Vẽ đường cong mất mát
plt.figure(figsize=(10, 6))
plt.plot(loss_history, color='green')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.grid()
plt.show()

# In hệ số cuối cùng
print(f"Final parameters: m = {m:.2f}, c = {c:.2f}")
```

### 2.3: Điều chỉnh tốc độ học 
• Nghiên cứu tác động của tốc độ học lên sự hội tụ. 
• Yêu cầu: 
1. Sử dụng f(x)=x2. 
2.  Triển  khai  phương  pháp  giảm  dần  độ  dốc  với  các  tốc  độ  học  khác  nhau 
(α=0,01,0,1,1,0,1,2). 
3. Vẽ đường cong hội tụ cho từng tốc độ học.

```python
import numpy as np
import matplotlib.pyplot as plt

# Hàm f_x
def f_x(x):
    return x**2

# Hàm Gradient
def grad(x):
    return 2*x

# Hàm GradientDescent
def myGD(alpha, x0, iter):
    x = [x0]
    for i in range(iter):
        x_new = x[-1] - alpha * grad(x[-1])
        if abs(grad(x_new)) < 1e-3:
            break
        x.append(x_new)
    return x, i+1

# Tốc độ học và giá trị khởi tạo
alpha = [0.01, 0.1, 1., 2.]  # tốc độ học
x0 = 10  # giá trị khởi tạo
iter = 100

# Tạo subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 6))  # Sửa plt.subplot thành plt.subplots

# Vẽ cho mỗi giá trị alpha trong từng subplot
for i, ax_i in zip(alpha, axes.flatten()):
    x_vals, iterations = myGD(i, x0, iter)
    f_x_vals = [f_x(x) for x in x_vals]
    ax_i.plot(f_x_vals, label=f'alpha = {i}')
    ax_i.set_title(f'Convergence for alpha = {i}')
    ax_i.set_xlabel('Iteration')
    ax_i.set_ylabel('f(x)')
    ax_i.grid(True, alpha=0.3)
    ax_i.legend()

# Hiển thị đồ thị
plt.tight_layout()
plt.show()
```

### 2.4:  Hồi  quy  logistic  với  Gradient  Descent  (Logistic  Regression  with  Gradient Descent) 
- Sử dụng gradient descent để huấn luyện mô hình hồi quy logistic. 
- Yêu cầu: 
1. Tạo dữ liệu phân loại nhị phân tổng hợp. 
2. Khởi tạo trọng số và độ lệch. 
3. Tính toán hàm sigmoid, gradient và độ lệch. 
4. Cập nhật trọng số và độ lệch theo từng bước. 

```python
import numpy as np

# 1. Tạo dữ liệu phân loại nhị phân tổng hợp
np.random.seed(42)
X = np.random.randn(100, 2)  # 100 mẫu, 2 đặc trưng
y = (np.dot(X, np.array([2, -1])) + 0.5 > 0).astype(int)  # Nhãn nhị phân

# 2. Khởi tạo trọng số và độ lệch
w = np.zeros(X.shape[1])
b = 0
learning_rate = 0.1
epochs = 1000
m = len(y)

# 3. Hàm sigmoid
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Hàm mất mát (binary cross-entropy loss)
def compute_loss(y, y_pred):
    return -np.mean(y * np.log(y_pred + 1e-8) + (1 - y) * np.log(1 - y_pred + 1e-8))

# Gradient descent
for epoch in range(epochs):
    # Tính z và dự đoán y
    z = np.dot(X, w) + b
    y_pred = sigmoid(z)
    
    # Tính gradient
    dw = np.dot(X.T, (y_pred - y)) / m
    db = np.sum(y_pred - y) / m
    
    # Cập nhật trọng số và độ lệch
    w -= learning_rate * dw
    b -= learning_rate * db
    
    # Tính toán và in hàm mất mát
    if epoch % 100 == 0:
        loss = compute_loss(y, y_pred)
        print(f"Epoch {epoch}, Loss: {loss:.4f}")

# 4. Hiển thị kết quả
print("\nTrọng số w:", w)
print("Độ lệch b:", b)

# Dự đoán trên dữ liệu huấn luyện
y_pred_classes = (sigmoid(np.dot(X, w) + b) > 0.5).astype(int)
accuracy = np.mean(y_pred_classes == y)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt

# 1. Vẽ biểu đồ phân tán của dữ liệu
plt.figure(figsize=(8, 6))
for label in np.unique(y):
    plt.scatter(X[y == label, 0], X[y == label, 1], label=f"Class {label}")

# 2. Vẽ đường phân cách
x_vals = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
y_vals = -(w[0] * x_vals + b) / w[1]  # Phương trình đường phân cách: w[0] * x + w[1] * y + b = 0
plt.plot(x_vals, y_vals, color='red', label='Decision Boundary')

# 3. Tùy chỉnh đồ thị
plt.title("Logistic Regression - Decision Boundary")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```

### 2.5: Stochastic Gradient Descent (SGD) 
- Triển khai gradient descent ngẫu nhiên trên một tập dữ liệu. 
- Yêu cầu: 
1. Sử dụng một tập dữ liệu đơn giản để hồi quy. 
2. Triển khai gradient descent theo lô và so sánh với SGD. 
3. Vẽ đường cong mất mát cho cả hai phương pháp.

```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo dữ liệu
np.random.seed(42)
X = np.random.rand(100, 1) * 10  # 100 điểm dữ liệu, 1 đặc trưng
y = 2 * X + 3 + np.random.randn(100, 1)  # Thêm nhiễu Gaussian

# Thêm cột bias
X_b = np.c_[np.ones((100, 1)), X]
# Khởi tạo tham số
theta = np.zeros((2, 1))
learning_rate = 0.01
epochs = 100
m = len(y)
losses_batch = []

# Batch Gradient Descent
for epoch in range(epochs):
    gradients = (2 / m) * X_b.T.dot(X_b.dot(theta) - y)
    theta -= learning_rate * gradients
    loss = (1 / m) * np.sum((X_b.dot(theta) - y) ** 2)
    losses_batch.append(loss)
theta_sgd = np.zeros((2, 1))
losses_sgd = []

# Stochastic Gradient Descent
for epoch in range(epochs):
    for i in range(m):
        random_index = np.random.randint(m)
        xi = X_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]
        gradients = 2 * xi.T.dot(xi.dot(theta_sgd) - yi)
        theta_sgd -= learning_rate * gradients
    loss = (1 / m) * np.sum((X_b.dot(theta_sgd) - y) ** 2)
    losses_sgd.append(loss)
plt.plot(losses_batch, label='Batch Gradient Descent')
plt.plot(losses_sgd, label='Stochastic Gradient Descent')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Loss Comparison: Batch vs. Stochastic Gradient Descent')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo tập dữ liệu đơn giản
np.random.seed(42)
X = np.random.rand(100, 1)
y = 3 * X + 2 + 0.5 * np.random.randn(100, 1)

# Thêm cột bias vào X
Xbar = np.c_[np.ones((X.shape[0], 1)), X]

# Hàm tính mất mát
def compute_loss(w, X, y):
    N = X.shape[0]
    return (1 / (2 * N)) * np.sum((X.dot(w) - y)**2)

# Hàm tính gradient (Batch Gradient Descent)
def batch_gradient(w, X, y):
    N = X.shape[0]
    return (1 / N) * X.T.dot(X.dot(w) - y)

# Hàm Gradient Descent theo lô (Batch Gradient Descent)
def batch_GD(w_init, X, y, eta, n_iters):
    w = [w_init]
    for _ in range(n_iters):
        grad = batch_gradient(w[-1], X, y)
        w_new = w[-1] - eta * grad
        w.append(w_new)
    return w

# Hàm tính gradient tại một điểm (SGD)
def sgrad(w, i, rd_id):
    true_i = rd_id[i]
    xi = Xbar[true_i, :]
    yi = y[true_i]
    a = np.dot(xi, w) - yi
    return (xi * a).reshape(2, 1)

# Hàm Stochastic Gradient Descent (SGD)
def SGD(w_init, X, y, eta, n_epochs):
    w = [w_init]
    w_last_check = w_init
    iter_check_w = 10
    N = X.shape[0]
    count = 0
    for _ in range(n_epochs):
        rd_id = np.random.permutation(N)  # Xáo trộn dữ liệu
        for i in range(N):
            count += 1
            g = sgrad(w[-1], i, rd_id)
            w_new = w[-1] - eta * g
            w.append(w_new)
            # Kiểm tra hội tụ
            if count % iter_check_w == 0:
                w_this_check = w_new
                if np.linalg.norm(w_this_check - w_last_check) / len(w_init) < 1e-3:
                    return w
                w_last_check = w_this_check
    return w

# Khởi tạo tham số
w_init = np.zeros((2, 1))
eta = 0.1
n_iters = 100
n_epochs = 10

# Chạy Batch Gradient Descent
w_batch = batch_GD(w_init, Xbar, y, eta, n_iters)
loss_batch = [compute_loss(w, Xbar, y) for w in w_batch]

# Chạy Stochastic Gradient Descent
w_sgd = SGD(w_init, Xbar, y, eta, n_epochs)
loss_sgd = [compute_loss(w, Xbar, y) for w in w_sgd]

# Vẽ đồ thị so sánh đường cong mất mát
plt.figure(figsize=(10, 6))
plt.plot(loss_batch, label='Batch Gradient Descent')
plt.plot(loss_sgd, label='Stochastic Gradient Descent', linestyle='dashed')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('Comparison of Batch GD and SGD')
plt.legend()
plt.show()

# In trọng số cuối cùng
print("Trọng số cuối cùng (Batch GD):", w_batch[-1].flatten())
print("Trọng số cuối cùng (SGD):", w_sgd[-1].flatten())
```

### 2.6: Giảm độ dốc theo lô nhỏ (Mini-Batch Gradient Descent) 
- Tối ưu hóa mô hình hồi quy bằng cách sử dụng giảm độ dốc theo lô nhỏ. 
- Yêu cầu: 
1. Chia tập dữ liệu thành các lô. 
2. Thực hiện cập nhật bằng cách sử dụng từng lô. 
3. So sánh tốc độ hội tụ của giảm độ dốc theo lô nhỏ và giảm độ dốc theo lô.

```python
# Mini-Batch Gradient Descent
batch_size = 20
theta_mini_batch = np.zeros((2, 1))
losses_mini_batch = []

for epoch in range(epochs):
    shuffled_indices = np.random.permutation(m)
    X_b_shuffled = X_b[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    
    for i in range(0, m, batch_size):
        xi = X_b_shuffled[i:i+batch_size]
        yi = y_shuffled[i:i+batch_size]
        gradients = (2 / batch_size) * xi.T.dot(xi.dot(theta_mini_batch) - yi)
        theta_mini_batch -= learning_rate * gradients
    loss = (1 / m) * np.sum((X_b.dot(theta_mini_batch) - y) ** 2)
    losses_mini_batch.append(loss)

# So sánh tốc độ hội tụ
plt.plot(losses_batch, label='Batch Gradient Descent')
plt.plot(losses_mini_batch, label='Mini-Batch Gradient Descent')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Loss Comparison: Batch vs. Mini-Batch Gradient Descent')
plt.show()
```

### 2.7: Động lượng trong Giảm độ dốc (Momentum in Gradient Descent) 
- Tăng cường giảm độ dốc bằng động lượng. 
- Yêu cầu: 
1. Triển khai giảm độ dốc có và không có động lượng. 
2. Sử dụng f(x)=x4−x2 để quan sát hiệu ứng. 
3. Vẽ quỹ đạo trong không gian tham số.

```python
def gradient_descent_with_momentum(f_prime, initial_x, lr=0.01, gamma=0.9, epochs=50):
    x = initial_x
    v = 0
    trajectory = [x]
    
    for _ in range(epochs):
        grad = f_prime(x)
        v = gamma * v + lr * grad
        x -= v
        trajectory.append(x)
    return trajectory

# Hàm f(x) = x^4 - x^2
f_prime = lambda x: 4 * x**3 - 2 * x
initial_x = 0.1
trajectory = gradient_descent_with_momentum(f_prime, initial_x)

# Vẽ quỹ đạo
x_vals = np.linspace(-1.5, 1.5, 100)
y_vals = x_vals**4 - x_vals**2
plt.plot(x_vals, y_vals, label='f(x)')
plt.plot(trajectory, [x**4 - x**2 for x in trajectory], 'ro-', label='Trajectory')
plt.legend()
plt.title('Gradient Descent with Momentum')
plt.show()
```

### 2.8: Tốc độ học thích ứng (Trình tối ưu hóa Adam) 
- Triển khai trình tối ưu hóa Adam. 
- Yêu cầu: 
1. Viết một triển khai tùy chỉnh của Adam. 
2. Sử dụng nó để giảm thiểu f(x,y)=x2+y2. 
3. So sánh hiệu suất của nó với gradient descent tiêu chuẩn.

```python
# Hàm triển khai Adam
def adam_optimizer(f_prime, initial_x, lr=0.01, beta1=0.9, beta2=0.999, epsilon=1e-8, epochs=100):
    x = initial_x
    m, v = 0, 0
    trajectory = [x]
    
    for t in range(1, epochs+1):
        grad = f_prime(x)
        m = beta1 * m + (1 - beta1) * grad
        v = beta2 * v + (1 - beta2) * grad**2
        m_hat = m / (1 - beta1**t)
        v_hat = v / (1 - beta2**t)
        x -= lr * m_hat / (np.sqrt(v_hat) + epsilon)
        trajectory.append(x)
    return trajectory

f_prime = lambda x: 2 * x
trajectory_adam = adam_optimizer(f_prime, initial_x)

# So sánh với gradient descent tiêu chuẩn
trajectory_gd = gradient_descent_with_momentum(f_prime, initial_x, lr=0.1, gamma=0)

# Vẽ quỹ đạo
plt.plot(trajectory_adam, label='Adam')
plt.plot(trajectory_gd, label='Standard Gradient Descent')
plt.legend()
plt.title('Comparison of Adam vs. Gradient Descent')
plt.show()
```

### 2.10: Sự hội tụ trong các hàm không lồi (Convergence in Non-Convex Functions) 
- Nghiên cứu những thách thức của gradient descent trong tối ưu hóa không lồi. 
- Yêu cầu: 
1. Sử dụng f(x)=x4−3x3+2. 
2. Triển khai gradient descent và trực quan hóa hành trình. 
3. Phân tích cực tiểu cục bộ so với cực tiểu toàn cục.

```python
# Hàm f(x) = x^4 - 3x^3 + 2
f = lambda x: x**4 - 3*x**3 + 2
f_prime = lambda x: 4*x**3 - 9*x**2
initial_x = 2
trajectory = gradient_descent_with_momentum(f_prime, initial_x, lr=0.01, gamma=0)

# Trực quan hóa
x_vals = np.linspace(-1, 3, 100)
y_vals = f(x_vals)
plt.plot(x_vals, y_vals, label='f(x)')
plt.plot(trajectory, [f(x) for x in trajectory], 'ro-', label='Trajectory')
plt.legend()
plt.title('Convergence in Non-Convex Function')
plt.show()
```
